```python
import networkx as nx
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']  # 设置中文字体
plt.rcParams['axes.unicode_minus'] = False

# 创建有向图
G = nx.DiGraph()

# 定义节点层级结构
nodes = {
    "第三方数据库": ["结构化数据", "半结构化数据", "非结构化数据"],
    "知识整合": ["实体抽取", "属性抽取"],
    "知识更新1": ["实体对齐", "实体消歧"],  # 处理重复节点
    "知识更新2": ["本体构建", "质量评估", "知识推理"],
    "知识图谱": ["应用服务"],
    "知识抽取": ["知识表示", "知识融合", "知识加工"]
}

# 添加节点和边
for parent, children in nodes.items():
    G.add_node(parent)
    for child in children:
        G.add_node(child)
        G.add_edge(parent, child)

# 设置专业布局坐标
pos = {
    # 第一层主节点
    "第三方数据库": (0, 4),
    "知识整合": (2, 4),
    "知识更新1": (4, 4),
    "知识更新2": (6, 4), 
    "知识图谱": (8, 4),
    "知识抽取": (10, 4),
    
    # 第二层子节点
    "结构化数据": (0, 3),
    "半结构化数据": (-1, 2), 
    "非结构化数据": (1, 2),
    "实体抽取": (2, 3),
    "属性抽取": (2, 2),
    "实体对齐": (4, 3),
    "实体消歧": (4, 2),
    "本体构建": (6, 3.5),
    "质量评估": (6, 3),
    "知识推理": (6, 2.5),
    "应用服务": (8, 3),
    "知识表示": (10, 3.5),
    "知识融合": (10, 3),
    "知识加工": (10, 2.5)
}

# 绘制图形
plt.figure(figsize=(15, 8))

# 绘制节点
nx.draw_networkx_nodes(G, pos, nodelist=nodes.keys(),
                      node_shape="s", node_size=5000,
                      node_color="lightblue", alpha=0.9)
nx.draw_networkx_nodes(G, pos, nodelist=[child for children in nodes.values() for child in children],
                      node_shape="o", node_size=3000,
                      node_color="lightgreen", alpha=0.8)

# 绘制边
nx.draw_networkx_edges(G, pos, arrows=True, 
                      arrowstyle="->", width=2,
                      connectionstyle="arc3,rad=0.1")

# 绘制标签
nx.draw_networkx_labels(G, pos, font_size=10, 
                       font_weight="bold", 
                       verticalalignment="center")

# 添加虚线框表示模块
for x in [0,2,4,6,8,10]:
    plt.gca().add_patch(plt.Rectangle((x-0.7, 3.5), 1.4, 1.2,
                        fill=False, linestyle="--", color="gray"))

plt.title("知识图谱技术架构示意图", fontsize=14, pad=20)
plt.xlim(-2, 12)
plt.ylim(1, 5)
plt.axis("off")
plt.tight_layout()
plt.show()
```
