```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# 
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial']
plt.rcParams['axes.unicode_minus'] = False

# 
dates = [datetime(2024,7,1), datetime(2024,8,1), datetime(2024,9,1)]
values = [400, 450, 430]  # 

plt.figure(figsize=(10, 6))
plt.plot(dates, values, marker='o', linestyle='-', color='green', linewidth=2)

# 
plt.annotate('Opening: 400 pts', xy=(dates[0], values[0]), 
             xytext=(dates[0], values[0]+10),  # 
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Mark-to-Market: 450 pts (+12.5%)', xy=(dates[1], values[1]), 
             xytext=(dates[1], values[1]-10),  # 
             arrowprops=dict(facecolor='red', shrink=0.05))
plt.annotate('Expiration: 430 pts (Total Return: +7.5%)', xy=(dates[2], values[2]), 
             xytext=(dates[2], values[2]+10),  # 
             arrowprops=dict(facecolor='blue', shrink=0.05))

# 
plt.title("Investment Performance (July 1 - Sept 1, 2024)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("APAC-PDI Index (Points)", fontsize=12)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=15))
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()    
```
