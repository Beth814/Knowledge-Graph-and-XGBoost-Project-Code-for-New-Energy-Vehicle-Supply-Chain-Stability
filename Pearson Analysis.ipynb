```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:


#计算相关系数热力图
import pandas as pd
import numpy  as np
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


t1=pd.read_excel("E:\\0 团队 知识工程\\副本相关指标.xlsx")

t2=t1.drop('年份',axis=1) #去掉年份变量


corr_matrix = t2.corr() #计算相关系数

plt.rcParams['font.sans-serif']=['SimHei']  #设置中文字体
plt.rcParams['axes.unicode_minus'] = False  #解决-号不能显示问题
# 使用Seaborn绘制热力图
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1,linewidths=0.5,square=True) #相关系数热力图

plt.title('Correlation Matrix')
plt.savefig('C:\\Users\\86136\\Desktop\\相关材料研究\\corr1.png')
plt.show()
#以上是计算相关系数热力图


# In[ ]:


#计算density和直方图
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置matplotlib的字体，以便支持中文和负号
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False


t1=pd.read_excel("E:\\0 团队 知识工程\\副本相关指标.xlsx")
data=t1.drop('年份',axis=1)
df = pd.DataFrame(data)

# 创建一个3行3列的子图布局
fig, axes = plt.subplots(3, 3, figsize=(18, 12))

# 遍历DataFrame的每一列，并在每个子图上绘制直方图和密度图
for i, col in enumerate(df.columns):
    row = i // 3
    col_idx = i % 3
    ax = axes[row, col_idx]
    
    # 绘制直方图
    sns.histplot(df[col], ax=ax, kde=False, color='red', stat="density", bins=30)
    
    # 绘制密度图
    sns.kdeplot(df[col], ax=ax, color='green', shade=True)
    ax.set_xlabel('Value') # 命名横坐标
    
    ax.set_title(f'{col}') # 设置标题
    
# 隐藏最后一行的第三个子图
axes[2, 2].set_visible(False)
# 调整子图间距
plt.tight_layout()
plt.savefig('C:\\Users\\86136\\Desktop\\相关材料研究\\stat1.png')
# 显示图表
plt.show()
#计算density和直方图

```
